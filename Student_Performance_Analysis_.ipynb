# **Data Preprocessing**

```python
#show only null values
data[data.isnull().any(axis=1)]
```

```python
data.duplicated().sum()
```

```python
# Handling Missing Values in different coulmns with mode valu
categorical_columns_with_missing = ['Parental_Education_Level', 'Distance_from_Home', 'Teacher_Quality']

for column in categorical_columns_with_missing:
    mode_value = data[column].mode()[0]
    data[column].fillna(mode_value, inplace=True)
data.sample(5)
```

```python
categorical_cols = data.select_dtypes(include=['object']).columns
print(categorical_cols)
```

```python
#convert categoical data into numerical data
df = data.copy()
for column in categorical_cols:
    unique_values = data[column].unique()
    mapping = {value: i for i, value in enumerate(unique_values)}
    df[column]= data[column].map(mapping)
df.head()
```

```python
# 1. Calculate correlations with 'Exam_Score'
correlations = df.corr()['Exam_Score']
sorted_correlations = correlations.sort_values(ascending=False)
print(sorted_correlations)
```

```python
#drop sum unnessery coulmns
data= data.drop(columns=['Peer_Influence'])
data= data.drop(columns=['Internet_Access'])
data= data.drop(columns=['Access_to_Resources'])
data= data.drop(columns=['Extracurricular_Activities'])
data= data.drop(columns=['Distance_from_Home'])
data.sample(5)
```

# **Data Visualization**

```python
# Correlation Heatmap for  only numerical features
numerical_features = data.select_dtypes(include=np.number)
correlation_matrix = numerical_features.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Purples', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# correlation with Exam_Score
correlation_with_target = numerical_features.corr()['Exam_Score'].sort_values(ascending=False)

plt.figure(figsize=(6, 6))
sns.barplot(x=correlation_with_target.index, y=correlation_with_target.values, palette='coolwarm')
plt.title('Correlation of Features with Exam Score')
plt.ylabel('Correlation Coefficient')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram for Exam Score
numerical_features.hist(bins=20, figsize=(10, 8))
plt.suptitle('Histogram of Numerical Features', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Box Plot for Exam Score
plt.figure(figsize=(8, 6))
sns.boxplot(x='Exam_Score', data=data, palette='Set3')
plt.title('Exam Score Distribution')
plt.show()
```

```python
# Removing outliers from dataset

Q1 = data['Exam_Score'].quantile(0.25)
Q3 = data['Exam_Score'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

data = data[(data['Exam_Score'] >= lower_bound) & (data['Exam_Score'] <= upper_bound)]
```

```python
# Box Plot for Exam Score
plt.figure(figsize=(8, 6))
sns.boxplot(x='Exam_Score', data=data, palette='Set3')
plt.title('Exam Score Distribution')
plt.show()
```

```python
#blox blot for numircal coulm
numerical_cols = data.select_dtypes(include=['number']).columns

for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=data[col])
    plt.title(f'Box Plots for: {col}')
    plt.show()
```

```python
## Calculate quantiles and IQR
Q1_tutoring = data['Tutoring_Sessions'].quantile(0.25)
Q3_tutoring = data['Tutoring_Sessions'].quantile(0.75)
IQR_tutoring = Q3_tutoring - Q1_tutoring

Q1_hours = data['Hours_Studied'].quantile(0.25)
Q3_hours = data['Hours_Studied'].quantile(0.75)
IQR_hours = Q3_hours - Q1_hours

# Calculate lower and upper bounds
lower_bound_tutoring = Q1_tutoring - 1.5 * IQR_tutoring
upper_bound_tutoring = Q3_tutoring + 1.5 * IQR_tutoring

lower_bound_hours = Q1_hours - 1.5 * IQR_hours
upper_bound_hours = Q3_hours + 1.5 * IQR_hours

# Filter data to remove outliers
data = data[(data['Tutoring_Sessions'] >= lower_bound_tutoring) & (data['Tutoring_Sessions'] <= upper_bound_tutoring)]
data = data[(data['Hours_Studied'] >= lower_bound_hours) & (data['Hours_Studied'] <= upper_bound_hours)]
```

```python
# Histogram for Exam Score
plt.figure(figsize=(6, 6))
sns.histplot(data['Exam_Score'], bins=20, kde=True)
plt.title("Histogram of Exam Scores")
plt.xlabel("Exam Score")
plt.ylabel("Frequency")
plt.show()
```

#**Feature Engineering**

```python
#Study Efficiency
data['Study_Efficiency'] = data['Hours_Studied'] * data['Attendance']
```

```python
#Wellness Index
data['Wellness_Index'] = (data['Sleep_Hours'] * data['Physical_Activity'])/2
```
